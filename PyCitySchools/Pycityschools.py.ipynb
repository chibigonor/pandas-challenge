```python
# passing math (the percentage of students who passed math)
passed_math_df = merge_df[merge_df['math_score'] > 80].count()
print (passed_math_df["student_name"])
```

```python
# passing reading (the percentage of students who passed reading)
passed_reading_df = merge_df[merge_df['reading_score'] > 80].count()
print (passed_reading_df["student_name"])
```

```python
# % passing reading (the percentage of students who passed reading)
percentage_passed_reading = (passed_reading_df/total_students)*100
print (percentage_passed_reading["student_name"])
```

```python
# % overall passing (the percentage of students who passed math AND reading)

passed_both_count = ((merge_df['math_score'] > 80) & (merge_df['reading_score'] > 80)).sum()
passed_both_count 
```

```python
passed_both_percent = (passed_both_count/total_students)*100
print (passed_both_percent)
```

```python
# School Summary
```

```python
grouped_schools = merge_df.groupby('school_name')
```

```python
summary_df = pd.DataFrame({
    'Total Students': grouped_schools['student_name'].count(),
    'Total School Budget': grouped_schools['budget'].mean(),
    'Per Student Budget': grouped_schools['budget'].mean() / grouped_schools['student_name'].count(),
    'Average Math Score': grouped_schools['math_score'].mean(),
    'Average Reading Score': grouped_schools['reading_score'].mean()})
summary_df
```

```python
# Calculate % passing math, reading, and overall
summary_df['% Passing Math'] = grouped_schools.apply(lambda x: (x['math_score'] >= 70).mean() * 100)
summary_df['% Passing Reading'] = grouped_schools.apply(lambda x: (x['reading_score'] >= 70).mean() * 100)
summary_df['% Overall Passing'] = grouped_schools.apply(lambda x: ((x['math_score'] >= 70) & (x['reading_score'] >= 70)).mean() * 100)

```

```python
summary_df['School Type'] = grouped_schools['type'].first()
```

```python
summary_df.reset_index(inplace=True)
```

```python
summary_df
```

```python
# Sort the schools by % Overall Passing in descending order
top_schools = summary_df.sort_values(by='% Overall Passing', ascending=False)

# Display the top 5 rows
top_schools = top_schools.head(5)

# Print the top schools DataFrame
top_schools
```

```python
# Sort the schools by % Overall Passing in ascending order
bottom_schools = summary_df.sort_values(by='% Overall Passing', ascending=True)

# Display the top 5 rows
bottom_schools = bottom_schools.head(5)

# Print the bottom schools DataFrame
bottom_schools
```

```python
# Group the original DataFrame by school name and grade
grouped_math_scores = merge_df.groupby(['school_name', 'grade'])

# Calculate the average math score for each group
average_math_scores = grouped_math_scores['math_score'].mean()

# Use pivot_table to reshape the data
math_scores_by_grade = average_math_scores.reset_index().pivot_table(index='school_name', columns='grade', values='math_score')

# Print the DataFrame
math_scores_by_grade
```

```python
# Group the original DataFrame by school name and grade
grouped_reading_scores = merge_df.groupby(['school_name', 'grade'])

# Calculate the average reading score for each group
average_reading_scores = grouped_reading_scores['reading_score'].mean()

# Use pivot_table to reshape the data
reading_scores_by_grade = average_reading_scores.reset_index().pivot_table(index='school_name', columns='grade', values='reading_score')

# Print the DataFrame
reading_scores_by_grade
```

```python
# Create bins and labels
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Slice the data and place it into bins
summary_df['Spending Ranges (Per Student)'] = pd.cut(summary_df['Per Student Budget'], bins=spending_bins, labels=labels)

# Group by spending ranges
spending_groups = summary_df.groupby('Spending Ranges (Per Student)')

# Calculate average math and reading scores, and percentages of passing
scores_by_spending = spending_groups[['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']].mean()

# Print the result
scores_by_spending
```

```python

spending_math_scores = scores_by_spending.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = scores_by_spending.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = scores_by_spending.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = scores_by_spending.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = scores_by_spending.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()

```

```python
# Create a DataFrame with the specified metrics
spending_summary = pd.DataFrame({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})

# Print the spending_summary DataFrame
spending_summary
```

```python
# Define the size bins and labels
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Bin the per_school_summary DataFrame based on school size
summary_df['School Size'] = pd.cut(summary_df['Total Students'], bins=size_bins, labels=labels)

# Group by school size
size_groups = summary_df.groupby('School Size')

# Calculate mean scores and percentages for each size category
size_math_scores = size_groups['Average Math Score'].mean()
size_reading_scores = size_groups['Average Reading Score'].mean()
size_passing_math = size_groups['% Passing Math'].mean()
size_passing_reading = size_groups['% Passing Reading'].mean()
size_overall_passing = size_groups['% Overall Passing'].mean()

# Create the size_summary DataFrame
size_summary = pd.DataFrame({
    "Average Math Score": size_math_scores,
    "Average Reading Score": size_reading_scores,
    "% Passing Math": size_passing_math,
    "% Passing Reading": size_passing_reading,
    "% Overall Passing": size_overall_passing
})

# Display the size_summary DataFrame
size_summary
```

```python
# Group by school type
type_groups = summary_df.groupby('School Type')

# Calculate mean scores and percentages for each type
type_math_scores = type_groups['Average Math Score'].mean()
type_reading_scores = type_groups['Average Reading Score'].mean()
type_passing_math = type_groups['% Passing Math'].mean()
type_passing_reading = type_groups['% Passing Reading'].mean()
type_overall_passing = type_groups['% Overall Passing'].mean()

# Create the type_summary DataFrame
type_summary = pd.DataFrame({
    "Average Math Score": type_math_scores,
    "Average Reading Score": type_reading_scores,
    "% Passing Math": type_passing_math,
    "% Passing Reading": type_passing_reading,
    "% Overall Passing": type_overall_passing
})

# Display the type_summary DataFrame
type_summary
```
